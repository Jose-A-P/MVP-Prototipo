```python
"""
Digital Twin Simulator para el MVP:
- Carga datos sintéticos (clientes_sinteticos.csv)
- Carga modelo y scaler si existen (modelo_logistico_impago.pkl, scaler_modelo.pkl)
- Si no existen, genera datos, clusteriza y entrena modelo rápido
- Aplica escenarios: caída ingresos, aumento tasas, peor comportamiento de pago, combinación
- Recalcula prob_impago y muestra agregados y cambios por cluster
"""

import os
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# PATH DE LOS ARCHIVOS UTILIZADOS
CSV_PATH = "clientes_sinteticos.csv"
SEG_CSV_OUT = "clientes_segmentados.csv"
MODEL_PATH = "modelo_logistico_impago.pkl"
SCALER_PATH = "scaler_modelo.pkl"
RANDOM_STATE = 42
```

```python
#Genera los datos sintenticos, en caso no se hayan generado en el entorno
def generar_datos_sinteticos(n=5000, save_csv=True):
    np.random.seed(RANDOM_STATE)
    segmentos = np.random.choice(
        ["bajo_riesgo", "medio_riesgo", "alto_riesgo"],
        size=n,
        p=[0.40, 0.35, 0.25]
    )

    score = np.zeros(n)
    ingresos = np.zeros(n)
    uso_credito = np.zeros(n)
    historial_impagos = np.zeros(n)
    frecuencia_pago = np.zeros(n)
    saldo_actual = np.zeros(n)
    antiguedad = np.random.uniform(0, 60, n)
    comportamiento_app = []

    for i, seg in enumerate(segmentos):
        if seg == "bajo_riesgo":
            score[i] = np.random.normal(780, 40)
            ingresos[i] = np.random.normal(15000, 3000)
            uso_credito[i] = np.random.beta(2, 10)
            historial_impagos[i] = np.random.poisson(0.1)
            frecuencia_pago[i] = np.random.normal(6, 2)
            saldo_actual[i] = np.random.gamma(2, 1500)
            comportamiento_app.append(np.random.choice(["alto", "medio"], p=[0.7, 0.3]))
        elif seg == "medio_riesgo":
            score[i] = np.random.normal(620, 60)
            ingresos[i] = np.random.normal(9000, 2500)
            uso_credito[i] = np.random.beta(3, 5)
            historial_impagos[i] = np.random.poisson(0.5)
            frecuencia_pago[i] = np.random.normal(4, 2)
            saldo_actual[i] = np.random.gamma(2, 2000)
            comportamiento_app.append(np.random.choice(["medio", "bajo"], p=[0.6, 0.4]))
        else:
            score[i] = np.random.normal(500, 80)
            ingresos[i] = np.random.normal(6000, 2000)
            uso_credito[i] = np.random.beta(5, 3)
            historial_impagos[i] = np.random.poisson(1.2)
            frecuencia_pago[i] = np.random.normal(2, 2)
            saldo_actual[i] = np.random.gamma(2, 3000)
            comportamiento_app.append("bajo")

    score = np.clip(score, 300, 900)
    ingresos = np.clip(ingresos, 1500, None)
    frecuencia_pago = np.clip(frecuencia_pago, 0, 12)

    logit = (
        -6
        + 0.003 * (650 - score)
        + 0.00020 * (8000 - ingresos)
        + 1.2 * uso_credito
        + 0.7 * historial_impagos
    )
    prob_impago = 1 / (1 + np.exp(-logit))
    impago = np.random.binomial(1, prob_impago)

    df = pd.DataFrame({
        "ID_cliente": [f"C{i:05d}" for i in range(n)],
        "segmento_oculto": segmentos,
        "score_crediticio": np.round(score, 2),
        "ingresos_mensuales": np.round(ingresos, 2),
        "uso_credito": np.round(uso_credito, 3),
        "historial_impagos": historial_impagos.astype(int),
        "frecuencia_pago": np.round(frecuencia_pago, 2),
        "saldo_actual": np.round(saldo_actual, 2),
        "antiguedad_cliente": np.round(antiguedad, 1),
        "comportamiento_app": comportamiento_app,
        "prob_impago": np.round(prob_impago, 4),
        "impago": impago
    })

    if save_csv:
        df.to_csv(CSV_PATH, index=False)
        print(f"[INFO] Dataset sintetico guardado en {CSV_PATH}")
    return df
```

```python
def train_model(df=None, k_optimo=3, save_csv=True):
    
    features = [
    "score_crediticio",
    "ingresos_mensuales",
    "uso_credito",
    "historial_impagos",
    "frecuencia_pago",
    "saldo_actual",
    "antiguedad_cliente"
    ]
    
    # Codificar variable categórica 
    df_encoded = pd.get_dummies(df[["comportamiento_app"]], drop_first=True)
    
    # Dataset final para clustering
    X = pd.concat([df[features], df_encoded], axis=1)  

    # ESCALAR DATOS
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    kmeans_final = KMeans(n_clusters=k_optimo, random_state=42)
    df["cluster_kmeans"] = kmeans_final.fit_predict(X_scaled)
    
    if save_csv:
        df.to_csv("clientes_segmentados.csv", index=False)
        print("\nArchivo generado: clientes_segmentados.csv")
    return df
```

```python
def load_or_train_model(df=None):

    features = [
        "score_crediticio",
        "ingresos_mensuales",
        "uso_credito",
        "historial_impagos",
        "frecuencia_pago",
        "saldo_actual",
        "antiguedad_cliente",
        "cluster_kmeans"
    ]

    # codificar comportamiento_app
    X = pd.concat([df[features], pd.get_dummies(df["comportamiento_app"], drop_first=True)], axis=1)
    y = df["impago"]

    # split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # entrenar
    model = LogisticRegression(max_iter=2000, class_weight="balanced", solver="lbfgs", random_state=RANDOM_STATE)
    model.fit(X_train_scaled, y_train)

    # exportar
    joblib.dump(model, MODEL_PATH)
    joblib.dump(scaler, SCALER_PATH)
    print(f"[INFO] Modelo entrenado y guardado en {MODEL_PATH} y {SCALER_PATH}")
    return model, scaler, X.columns
```

```python
# PREDICCION CON MODELO

def predict_prob(model, scaler, df, feature_cols):
    # Se construye X utilizando las mismas columnas del entrenamiento del modelo
    X = pd.concat([df[feature_cols], pd.get_dummies(df["comportamiento_app"], drop_first=True)], axis=1)
    
    # Si faltan columnas, le agrega 0s
    for col in feature_cols:
        if col not in X.columns:
            X[col] = 0
            
    # Reindex que asegura orden de los datos
    X = X.reindex(columns=feature_cols + [c for c in X.columns if c not in feature_cols], fill_value=0)
    X_scaled = scaler.transform(X)
    probs = model.predict_proba(X_scaled)[:, 1]
    return probs
```

```python
# ESCENARIOS
def scenario_income_drop(df, pct_drop=0.2):
    """Reduce ingresos en pct_drop """
    df2 = df.copy()
    df2["ingresos_mensuales"] = df2["ingresos_mensuales"] * (1 - pct_drop)
    return df2

def scenario_interest_rate_increase(df, extra_rate=0.03):
    """
    Aumenta 'costo' del crédito:
    - incrementa saldo estimado proporcional al extra_rate.
    - Se asume que mayor tasa aumenta la carga financiera y afecta saldo/uso_credito:
    """
    df2 = df.copy()
    # incrementar saldo actual en factor relacionado a extra_rate
    df2["saldo_actual"] = df2["saldo_actual"] * (1 + extra_rate * 2)  # factor heurístico
    # también aumentar uso_credito ligeramente
    df2["uso_credito"] = (df2["uso_credito"] + extra_rate * 3).clip(0, 1)
    return df2

def scenario_worse_payment_behavior(df, extra_missed=0.5):
    """
    Simula peor comportamiento: aumento en historial_impagos y reducción de frecuencia de pago.
    """
    df2 = df.copy()
    df2["historial_impagos"] = (df2["historial_impagos"] + np.random.poisson(lam=extra_missed, size=len(df2))).astype(int)
    df2["frecuencia_pago"] = (df2["frecuencia_pago"] - extra_missed).clip(0)
    return df2

def scenario_combined(df, income_drop=0.2, rate_inc=0.03, extra_missed=0.5):
    """
    Simula en conjunto los 3 posibles escenarios
    """
    df2 = scenario_income_drop(df, income_drop)
    df2 = scenario_interest_rate_increase(df2, rate_inc)
    df2 = scenario_worse_payment_behavior(df2, extra_missed)
    return df2
```

```python
# RESUMEN DE LA BASE Y EL ESCENARIO, JUNTO A LAS METRICAS DE SU COMPORTAMIENTO 
def summarize_impact(df_base, df_scenario, probs_base, probs_scenario, top_n=10):
    df_res = df_base[["ID_cliente", "cluster_kmeans"]].copy()
    df_res["prob_base"] = probs_base
    df_res["prob_scenario"] = probs_scenario
    df_res["delta_prob"] = df_res["prob_scenario"] - df_res["prob_base"]
    
    # agregados
    overall = {
        "mean_prob_base": df_res["prob_base"].mean(),
        "mean_prob_scenario": df_res["prob_scenario"].mean(),
        "mean_delta": df_res["delta_prob"].mean(),
        "pct_increase_mean": (df_res["prob_scenario"].mean() - df_res["prob_base"].mean()) / max(df_res["prob_base"].mean(), 1e-9)
    }
    
    by_cluster = df_res.groupby("cluster_kmeans").agg({
        "prob_base": "mean",
        "prob_scenario": "mean",
        "delta_prob": "mean",
        "ID_cliente": "count"
    }).rename(columns={"ID_cliente": "n_clients"})
    
    top_risers = df_res.sort_values("delta_prob", ascending=False).head(top_n)
    return df_res, overall, by_cluster, top_risers

# PLOTEO DE LA DISTRIBUCION DE PROBABILIDAD BASE Y ESCENARIO
def plot_prob_distributions(df_res, title="Distribución de probabilidad"):
    plt.figure(figsize=(10,5))
    sns.histplot(df_res["prob_base"], kde=True, label="Base", stat="density", color="C0", alpha=0.5)
    sns.histplot(df_res["prob_scenario"], kde=True, label="Escenario", stat="density", color="C1", alpha=0.5)
    plt.legend()
    plt.title(title)
    plt.xlabel("Probabilidad de impago")
    plt.show()

# PLOTEO DEL DELTA EN LA PROBABILIDAD DE IMPAGO POR CLUSTER
def plot_delta_by_cluster(by_cluster):
    by_cluster = by_cluster.reset_index()
    plt.figure(figsize=(8,4))
    sns.barplot(x="cluster_kmeans", y="delta_prob", data=by_cluster, palette="coolwarm")
    plt.title("Cambio medio en probabilidad por cluster")
    plt.xlabel("Cluster")
    plt.ylabel("Delta probabilidad")
    plt.show()
```

```python
#FLUJO DE EJEMPLO PARA LOS POSIBLES ESCENARIOS
def run_demo(do_export=False):
    
    # cargar o generar datos
    if os.path.exists(CSV_PATH):
        df = pd.read_csv(CSV_PATH)
        print(f"[INFO] Cargado {CSV_PATH}")
    else:
        print("[WARN] No se encontró dataset. Generando dataset sintético...")
        df = generar_datos_sinteticos(n=5000, save_csv=True)

    if "cluster_kmeans" not in df.columns:
        # crear clusters utilizando modelo original
        df = train_model(df, k_optimo=6, save_csv=True)
        print(f"[INFO] Se creo cluster_kmeans y se guardo {SEG_CSV_OUT}")

    # cargar modelo y scaler si existen, sino entrenar
    if os.path.exists(MODEL_PATH) and os.path.exists(SCALER_PATH):
        model = joblib.load(MODEL_PATH)
        scaler = joblib.load(SCALER_PATH)
        
        base_features = [
            "score_crediticio",
            "ingresos_mensuales",
            "uso_credito",
            "historial_impagos",
            "frecuencia_pago",
            "saldo_actual",
            "antiguedad_cliente",
            "cluster_kmeans"
        ]

        feature_cols = base_features.copy()
        print(f"[INFO] Modelo cargado desde {MODEL_PATH}")
    else:
        print("[WARN] No se encontró modelo pre-entrenado. Entrenando modelo rápido con los datos disponibles...")
        model, scaler, feature_cols = load_or_train_model(df)

    # predecir probabilidad de impago original (linea base)
    probs_base = predict_prob(model, scaler, df, feature_cols)

    # seleccionar escenario y simular 
    df_income = scenario_income_drop(df, pct_drop=0.2)
    df_rate = scenario_interest_rate_increase(df, extra_rate=0.1)
    df_behavior = scenario_worse_payment_behavior(df, extra_missed=0.7)
    df_combined = scenario_combined(df, income_drop=0.2, rate_inc=0.03, extra_missed=0.7)

    # predecir probabilidades bajo escenarios
    probs_inc = predict_prob(model, scaler, df_income, feature_cols)
    probs_rate = predict_prob(model, scaler, df_rate, feature_cols)
    probs_beh = predict_prob(model, scaler, df_behavior, feature_cols)
    probs_comb = predict_prob(model, scaler, df_combined, feature_cols)
    
    # ------------------------------------------------
    # Resume el impacto del escenario simulado, permitiendo observar su impacto sobre todo el dataset y los
    # clientes a los que mas ha impactado. 
    # Modificar el ultimo parametro de la funcion siguiente por uno de los probs previos, para observar su comportamiento.
    # ------------------------------------------------
    df_res, overall, by_cluster, top_risers = summarize_impact(df, df_combined, probs_base, probs_beh)

    print("\n=== Resumen general (escenario combinado) ===")
    for k,v in overall.items():
        print(f"{k}: {v}")

    print("\n=== Impacto por cluster ===")
    print(by_cluster)

    print("\n=== Top clientes con mayor aumento en probabilidad ===")
    print(top_risers.head(10))

    # 7) plots
    plot_prob_distributions(df_res, title="Base vs Escenario combinado (prob impago)")
    plot_delta_by_cluster(by_cluster)

    # 8) exportar
    if do_export:
        df_out = df.copy()
        df_out["prob_base"] = probs_base
        df_out["prob_combined"] = probs_comb
        df_out.to_csv("clientes_simulados_combined.csv", index=False)
        df_res.to_csv("impacto_combined_por_cliente.csv", index=False)
        print("[INFO] Exportados clientes_simulados_combined.csv e impacto_combined_por_cliente.csv")

    return {
        "df_base": df,
        "df_combined": df_combined,
        "probs_base": probs_base,
        "probs_combined": probs_comb,
        "summary": overall,
        "by_cluster": by_cluster,
        "top_risers": top_risers
    }
```

```python
resultado = run_demo(do_export=True)
print("[DONE] Simulación completada.")
```
