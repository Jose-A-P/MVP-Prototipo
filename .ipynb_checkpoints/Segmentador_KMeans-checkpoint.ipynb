```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# CARGAR DATA
df = pd.read_csv("clientes_sinteticos.csv")

# Variables numéricas para clustering
features = [
    "score_crediticio",
    "ingresos_mensuales",
    "uso_credito",
    "historial_impagos",
    "frecuencia_pago",
    "saldo_actual",
    "antiguedad_cliente"
]

# Codificar variable categórica 
df_encoded = pd.get_dummies(df[["comportamiento_app"]], drop_first=True)

# Dataset final para clustering
X = pd.concat([df[features], df_encoded], axis=1)
```

```python
# ESCALAR DATOS
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# DETERMINAR NÚMERO ÓPTIMO DE CLUSTERS (ELBOW)

inertia = []
K = range(2, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(7,4))
plt.plot(K, inertia, marker='o')
plt.title("Método del Codo (Elbow Method)")
plt.xlabel("Número de clusters (k)")
plt.ylabel("Inercia (Distortion)")
plt.grid(True)
plt.show()
```

```python
# ENTRENAR KMEANS FINAL

k_optimo = 6

kmeans_final = KMeans(n_clusters=k_optimo, random_state=42)
df["cluster_kmeans"] = kmeans_final.fit_predict(X_scaled)
```

```python
# PCA PARA VISUALIZAR CLUSTERS EN 2D

pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)

df["PCA1"] = pca_result[:, 0]
df["PCA2"] = pca_result[:, 1]

plt.figure(figsize=(8,6))
sns.scatterplot(
    x="PCA1",
    y="PCA2",
    hue="cluster_kmeans",
    palette="Set2",
    data=df,
    alpha=0.7
)
plt.title("Clusters KMeans visualizados con PCA (2D)")
plt.show()
```

```python
# ANÁLISIS DE PERFILES DE CLUSTER

cluster_summary = df.groupby("cluster_kmeans")[features + ["prob_impago"]].mean()
print("\nResumen de clusters:\n")
print(cluster_summary)

print("\nDistribución de clientes por cluster:")
print(df["cluster_kmeans"].value_counts())
```

```python
# GUARDAR DATASET CON CLUSTERS
df.to_csv("clientes_segmentados.csv", index=False)
print("\nArchivo generado: clientes_segmentados.csv")
```
