```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    f1_score,
    confusion_matrix,
    RocCurveDisplay
)
import matplotlib.pyplot as plt
import seaborn as sns

# CARGAR DATA GENERADA CON KMEANS
df = pd.read_csv("clientes_segmentados.csv")
```

```python
# SELECCIONAR FEATURES Y TARGET
features = [
    "score_crediticio",
    "ingresos_mensuales",
    "uso_credito",
    "historial_impagos",
    "frecuencia_pago",
    "saldo_actual",
    "antiguedad_cliente",
    "cluster_kmeans"
]

# One-hot de comportamiento_app
df_encoded = pd.get_dummies(df["comportamiento_app"], drop_first=True)

# X final
X = pd.concat([df[features], df_encoded], axis=1)

# Target
y = df["impago"]
```

```python
# SPLIT TRAIN/TEST
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# ESCALAR VARIABLES NUMRICAS
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# ENTRENAR REGRESIN LOGSTICA
logreg = LogisticRegression(
    max_iter=2000,
    class_weight="balanced",
    solver="lbfgs"
)
logreg.fit(X_train_scaled, y_train)

# PREDICCIONES
y_pred = logreg.predict(X_test_scaled)
y_prob = logreg.predict_proba(X_test_scaled)[:, 1]
```

```python
# MTRICAS DEL MODELO
accuracy = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("\n Resultados del Modelo de Regresi贸n Log铆stica")
print(f"Accuracy: {accuracy:.4f}")
print(f"AUC: {auc:.4f}")
print(f"F1 Score: {f1:.4f}")
print("\nMatriz de confusi贸n:\n", cm)
```

```python
# CURVA ROC
plt.figure(figsize=(6,5))
RocCurveDisplay.from_predictions(y_test, y_prob)
plt.title("Curva ROC - Modelo de Impago")
plt.grid(True)
plt.show()
```

```python
# IMPORTANCIA DE VARIABLES
coef = pd.DataFrame({
    "feature": X.columns,
    "coeficiente": logreg.coef_[0]
}).sort_values(by="coeficiente", ascending=False)

plt.figure(figsize=(8, 6))
sns.barplot(data=coef, x="coeficiente", y="feature", palette="viridis")
plt.title("Importancia de Variables (Regresi贸n Log铆stica)")
plt.show()

print("\nImportancia de features:\n")
print(coef)
```

```python
# EXPORTAR MODELO
import joblib
joblib.dump(logreg, "modelo_logistico_impago.pkl")
joblib.dump(scaler, "scaler_modelo.pkl")

print("\nModelo exportado como: modelo_logistico_impago.pkl")
print("Scaler exportado como: scaler_modelo.pkl")
```
